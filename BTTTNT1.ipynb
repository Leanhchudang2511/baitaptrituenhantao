```python

import folium
from folium.plugins import MiniMap, Fullscreen
from math import radians, sin, cos, sqrt, atan2

CENTER = (10.7769, 106.6953)

def khoang_cach(lat1, lon1, lat2, lon2):
    R = 6371
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    return round(R * 2 * atan2(sqrt(a), sqrt(1 - a)), 2)

PLACES = [
    ("hospital", "red",    "plus-sign",     "Bệnh viện Chợ Rẫy",              10.7553, 106.6680),
    ("hospital", "red",    "plus-sign",     "Bệnh viện ĐH Y Dược TP.HCM",     10.7620, 106.6822),
    ("hospital", "red",    "plus-sign",     "Bệnh viện Nhân Dân 115",          10.7744, 106.6670),
    ("mall",     "blue",   "shopping-cart", "Vincom Center Bà Triệu",          10.7741, 106.7015),
    ("mall",     "blue",   "shopping-cart", "Chợ Bến Thành",                   10.7717, 106.6980),
    ("mall",     "blue",   "shopping-cart", "Nowzone Fashion Mall",             10.7589, 106.6889),
    ("mall",     "blue",   "shopping-cart", "Co.opmart Lý Thường Kiệt",        10.7714, 106.6721),
    ("gov",      "green",  "home",          "UBND Quận 5",                     10.7526, 106.6693),
    ("gov",      "green",  "home",          "UBND Quận 10",                    10.7748, 106.6681),
    ("gov",      "green",  "home",          "Bến xe Miền Tây",                 10.7376, 106.6282),
    ("other",    "orange", "leaf",          "Công viên Lê Thị Riêng",          10.7859, 106.6847),
    ("other",    "orange", "leaf",          "Bảo tàng Chứng tích Chiến tranh", 10.7796, 106.6921),
]

m = folium.Map(location=CENTER, zoom_start=14)
MiniMap().add_to(m)
Fullscreen().add_to(m)

layers = {
    "hospital": folium.FeatureGroup(name="🏥 Bệnh viện",        show=True),
    "mall":     folium.FeatureGroup(name="🛒 Thương mại / Chợ", show=True),
    "gov":      folium.FeatureGroup(name="🏛 Cơ quan hành chính",show=True),
    "other":    folium.FeatureGroup(name="📍 Khác",              show=True),
}

folium.Marker(
    CENTER,
    popup=folium.Popup("<b>⭐ UEH – Đại học Kinh tế TP.HCM</b><br>📍 Điểm trung tâm", max_width=250),
    tooltip="⭐ UEH – Trung tâm",
    icon=folium.Icon(color="purple", icon="star", prefix="glyphicon")
).add_to(m)

for group, color, icon, name, lat, lon in PLACES:
    dist = khoang_cach(CENTER[0], CENTER[1], lat, lon)
    popup_html = (
        f"<b>{name}</b><br>"
        f"📍 {lat}, {lon}<br>"
        f"📏 Cách UEH: <b>{dist} km</b>"
    )
    folium.Marker(
        (lat, lon),
        popup=folium.Popup(popup_html, max_width=260),
        tooltip=name,
        icon=folium.Icon(color=color, icon=icon, prefix="glyphicon")
    ).add_to(layers[group])

for layer in layers.values():
    layer.add_to(m)

folium.LayerControl(collapsed=False).add_to(m)

m
```
