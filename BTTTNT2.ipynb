```python
!pip install geopandas folium -q

import geopandas as gpd
import folium
import pandas as pd
from folium import Choropleth, GeoJson, GeoJsonTooltip
from IPython.display import display
from google.colab import drive

drive.mount('/content/drive')
gdf = gpd.read_file("/content/drive/MyDrive/vnm_admin1.geojson")


dan_so = {
    "An Giang": 1908, "Bac Lieu": 910, "Bac Giang": 1803,
    "Bac Kan": 313, "Bac Ninh": 1421, "Ben Tre": 1288,
    "Binh Dinh": 1497, "Binh Duong": 2593, "Binh Phuoc": 1017,
    "Binh Thuan": 1230, "Ca Mau": 1194, "Can Tho": 1235,
    "Cao Bang": 537, "Da Nang": 1134, "Dak Lak": 1869,
    "Dak Nong": 658, "Dien Bien": 633, "Dong Nai": 3231,
    "Dong Thap": 1590, "Gia Lai": 1513, "Ha Giang": 875,
    "Ha Nam": 863, "Ha Noi": 8246, "Ha Tinh": 1296,
    "Hai Duong": 1916, "Hai Phong": 2028, "Hau Giang": 744,
    "Hoa Binh": 863, "Hung Yen": 1252, "Khanh Hoa": 1232,
    "Kien Giang": 1870, "Kon Tum": 566, "Lai Chau": 492,
    "Lam Dong": 1296, "Lang Son": 782, "Lao Cai": 757,
    "Long An": 1698, "Nam Dinh": 1780, "Nghe An": 3327,
    "Ninh Binh": 945, "Ninh Thuan": 621, "Phu Tho": 1463,
    "Phu Yen": 967, "Quang Binh": 884, "Quang Nam": 1493,
    "Quang Ngai": 1237, "Quang Ninh": 1320, "Quang Tri": 638,
    "Soc Trang": 1195, "Son La": 1248, "Tay Ninh": 1169,
    "Thai Binh": 1786, "Thai Nguyen": 1278, "Thanh Hoa": 3640,
    "Thua Thien Hue": 1128, "Tien Giang": 1766, "Ho Chi Minh City": 9166,
    "Tra Vinh": 1009, "Tuyen Quang": 793, "Vinh Long": 1023,
    "Vinh Phuc": 1151, "Yen Bai": 830,
}

df = pd.DataFrame(list(dan_so.items()), columns=["adm1_name", "dan_so"])
gdf = gdf.merge(df, on="adm1_name", how="left")
gdf["dan_so"] = gdf["dan_so"].fillna(0)

gdf_plot = gdf[["adm1_name", "adm1_name1", "dan_so", "geometry"]].copy()


m = folium.Map(location=[16.0, 106.0], zoom_start=6, tiles="CartoDB positron")

Choropleth(
    geo_data=gdf_plot.__geo_interface__,
    data=gdf_plot,
    columns=["adm1_name", "dan_so"],
    key_on="feature.properties.adm1_name",
    fill_color="YlOrRd",
    fill_opacity=0.75,
    line_opacity=0.3,
    legend_name="Dân số (nghìn người)",
    nan_fill_color="lightgrey",
).add_to(m)

GeoJson(
    gdf_plot.__geo_interface__,
    tooltip=GeoJsonTooltip(
        fields=["adm1_name1", "dan_so"],
        aliases=["Tỉnh/Thành:", "Dân số (nghìn):"],
    ),
    style_function=lambda x: {"fillOpacity": 0, "weight": 0},
).add_to(m)

display(m)
```
