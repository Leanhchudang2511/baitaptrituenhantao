```python

import time
import folium
from folium.plugins import MiniMap, Fullscreen
from geopy.geocoders import Nominatim
from geopy.distance import geodesic

CENTER_ADDR = "Đại học Kinh tế TP.HCM, 59C Nguyễn Đình Chiểu, Quận 3, TP.HCM"
CENTER      = (10.7769, 106.6953)
CENTER_NAME = "UEH – Đại học Kinh tế TP.HCM"

ADDRESSES = [
    "Bệnh viện Chợ Rẫy, 201B Nguyễn Chí Thanh, Quận 5, TP.HCM",
    "Bệnh viện Nhân Dân 115, 527 Sư Vạn Hạnh, Quận 10, TP.HCM",
    "Bệnh viện Đại học Y Dược, 215 Hồng Bàng, Quận 5, TP.HCM",
    "Chợ Bến Thành, Quận 1, TP.HCM",
    "Vincom Center, 72 Lê Thánh Tôn, Quận 1, TP.HCM",
    "Co.opmart Lý Thường Kiệt, Quận 10, TP.HCM",
    "UBND Quận 5, 2 Trần Hưng Đạo, Quận 5, TP.HCM",
    "UBND Quận 10, 600 Nguyễn Tri Phương, Quận 10, TP.HCM",
    "Bến xe Miền Tây, Bình Chánh, TP.HCM",
    "Công viên Lê Thị Riêng, Quận 10, TP.HCM",
    "Bảo tàng Chứng tích Chiến tranh, 28 Võ Văn Tần, Quận 3, TP.HCM",
    "Nhà thờ Đức Bà, Quận 1, TP.HCM",
]

geolocator = Nominatim(user_agent="bai232_ueh", timeout=10)

def geocode(address):
    try:
        loc = geolocator.geocode(address)
        if loc:
            return (loc.latitude, loc.longitude)
    except Exception as e:
        print(f"  ⚠ Lỗi: {address} → {e}")
    return None

print("🔍 Đang geocode địa chỉ...\n")
results = []
for addr in ADDRESSES:
    coords = geocode(addr)
    time.sleep(1)
    if coords:
        dist = round(geodesic(CENTER, coords).km, 2)
        results.append({"name": addr.split(",")[0], "addr": addr, "coords": coords, "dist": dist})
        print(f"  ✓ {addr.split(',')[0]:40s} → {coords}  |  {dist} km")
    else:
        print(f"  ✗ Không tìm thấy: {addr}")

print(f"\n✅ Geocode thành công: {len(results)}/{len(ADDRESSES)} địa chỉ")

m = folium.Map(location=CENTER, zoom_start=13)
MiniMap().add_to(m)
Fullscreen().add_to(m)

folium.Marker(
    CENTER,
    popup=folium.Popup(f"<b>⭐ {CENTER_NAME}</b>", max_width=250),
    tooltip="⭐ UEH – Trung tâm",
    icon=folium.Icon(color="purple", icon="star", prefix="glyphicon")
).add_to(m)

def mau_theo_kc(dist):
    if dist < 1:   return "green"
    if dist < 3:   return "blue"
    if dist < 6:   return "orange"
    return "red"

for p in results:
    color = mau_theo_kc(p["dist"])

    folium.PolyLine(
        locations=[CENTER, p["coords"]],
        color=color,
        weight=2,
        opacity=0.7,
        tooltip=f"{p['name']} – {p['dist']} km"
    ).add_to(m)

    popup_html = (
        f"<b>{p['name']}</b><br>"
        f"📍 {p['addr']}<br>"
        f"🌐 {p['coords'][0]:.5f}, {p['coords'][1]:.5f}<br>"
        f"📏 Cách UEH: <b>{p['dist']} km</b>"
    )
    folium.Marker(
        p["coords"],
        popup=folium.Popup(popup_html, max_width=300),
        tooltip=f"{p['name']} ({p['dist']} km)",
        icon=folium.Icon(color=color, icon="map-marker", prefix="glyphicon")
    ).add_to(m)

legend = """
<div style="position:fixed; bottom:30px; left:30px; z-index:1000;
     background:white; padding:10px 14px; border-radius:8px;
     border:1px solid #ccc; font-size:13px; line-height:1.8">
  <b>📏 Khoảng cách đến UEH</b><br>
  <span style="color:green">●</span> &lt; 1 km<br>
  <span style="color:blue">●</span> 1 – 3 km<br>
  <span style="color:orange">●</span> 3 – 6 km<br>
  <span style="color:red">●</span> &gt; 6 km
</div>
"""
m.get_root().html.add_child(folium.Element(legend))

m
```
